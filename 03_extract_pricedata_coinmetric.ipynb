```python
import urllib.request
import os.path
if not os.path.isfile('./data/coinmetric_btc.csv'):
    print("./data/coinmetric_btc.csv is not existing!")
    print("Download from https://coinmetrics.io/community-network-data/#comm-files (Bitcoin)")
    print("rename and place it in the data-folder")
```

```python
import pandas as pd
btc = pd.read_csv("./data/coinmetric_btc.csv",\
                  index_col="date",parse_dates=["date"])
btc
```

```python
daily_prices = pd.DataFrame()
daily_prices['price'] = btc.PriceUSD.resample('M').mean()
btc = daily_prices
# the first year is not very reliable, we'll skip it.
#btc = btc[btc.index > '2012-1-1']
# Do we really need that?!
# vielleicht besser btc.notnull()
btc.fillna(method='ffill',inplace=True) # replace Null-values with forwardfill
print("So now we have {} datapoints".format(btc.shape[0]))
btc
```

```python
btc.to_csv ('data/coinmetric_btc_monthly.csv', index = True, header=True)
```
