# Load all the blocks


```python
import glob
import pandas as pd

header = ['height','subsidy','hash','strippedsize','size','weight','version','versionHex','merkleroot','time','mediantime','nonce','bits','difficulty','chainwork','nTx','previousblockhash','nextblockhash']
blockdata = pd.DataFrame(columns = header)
blockdata.set_index("height")
for file in glob.glob('data/blocks_*.csv'):
    blockdata_chunk = pd.read_csv(file)
    blockdata_chunk.columns = header
    blockdata_chunk.set_index("height")
    # effectively appending the newly loaded Dataframe to the big one
    blockdata = pd.concat([blockdata, blockdata_chunk])
    print("processed {}".format(file))
blockdata
```

```python
blockdata["date"]= pd.to_datetime(blockdata.time,unit='s')
blockdata = blockdata.set_index("date")
blockdata
```

```python
monthly_flow = pd.DataFrame()
monthly_flow['flow'] = blockdata.subsidy.resample("M").sum()
monthly_flow
```

```python
monthly_flow.to_csv('./data/flow_monthly.csv')
```
